```python
import tensorflow as tf
from keras import layers, models

# Define the CNN model
def create_cnn_model(input_shape):
    model = models.Sequential()

    # Convolutional layers
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
    model.add(layers.MaxPooling2D((2, 2)))

    # Flatten layer to transition from convolutional to fully connected layers
    model.add(layers.Flatten())

    # Fully connected layers
    model.add(layers.Dense(128, activation='relu'))
    model.add(layers.Dense(2, activation='softmax'))  # Assuming binary classification

    return model

# Assuming your input spectrogram shape is (height, width, channels)
input_shape = (15, 15, 1)  # Adjust dimensions based on your actual input shape

# Create the model
model = create_cnn_model(input_shape)

# Display the model summary
model.summary()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
ke = 8.99e9  # Coulomb's constant in N m^2/C^2
q1 = 1e-6   # Charge 1 in Coulombs
q2 = 1e-6   # Charge 2 in Coulombs
lambda_ = 1e-6  # Linear charge density in C/m

# Distance range
r = np.linspace(0.01, 1, 400)  # From 0.01 m to 1 m to avoid division by zero

# Electrostatic force between point charges
F = ke * q1 * q2 / r**2

# Electric field due to a point charge
E_point = ke * q1 / r**2

# Electric field due to an infinitely long thin wire
E_wire = 2 * ke * lambda_ / r

# Plotting
plt.figure(figsize=(12, 8))

# Electrostatic Force
plt.subplot(3, 1, 1)
plt.plot(r, F, label=r'$F = k_e \frac{q_1 q_2}{r^2}$')
plt.title('Electrostatic Force between Point Charges vs. Distance')
plt.xlabel('Distance (m)')
plt.ylabel('Force (N)')
plt.legend()
plt.grid(True)

# Electric Field due to a Point Charge
plt.subplot(3, 1, 2)
plt.plot(r, E_point, label=r'$E = k_e \frac{q}{r^2}$', color='orange')
plt.title('Electric Field due to a Point Charge vs. Distance')
plt.xlabel('Distance (m)')
plt.ylabel('Electric Field (N/C)')
plt.legend()
plt.grid(True)

# Electric Field due to an Infinitely Long Thin Wire
plt.subplot(3, 1, 3)
plt.plot(r, E_wire, label=r'$E = \frac{2k_e \lambda}{r}$', color='green')
plt.title('Electric Field due to an Infinitely Long Thin Wire vs. Distance')
plt.xlabel('Distance (m)')
plt.ylabel('Electric Field (N/C)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
